```python
# Concatenate all Q2 values into a single array
q2_values_1 = q2reg1_mc_ele_signal['q2']
q2_values_2 = q2reg2_mc_ele_signal['q2']
q2_values_3 = q2reg3_mc_ele_signal['q2']
q2_values_4 = q2reg4_mc_ele_signal['q2']

# Combine all Q2 values
q2_values = np.concatenate([q2_values_1, q2_values_2, q2_values_3, q2_values_4])

# Remove NaN values
q2_values = q2_values[~np.isnan(q2_values)]

# Plot histogram with histtype='stepfilled'
plt.hist(q2_values, bins=100, color='b', density=True, histtype='stepfilled', alpha=0.75)
plt.title('Q2 regions for electron')
plt.xlabel(r'$q^2$ (GeV$^2$)')
plt.ylabel('Probability Density')

# Add vertical dashed lines to separate the Q2 regions
q2_regions = [q2_values_1.min(), q2_values_1.max(), q2_values_2.max(), q2_values_3.max(), q2_values_4.max()]
for i in range(len(q2_regions)-1):
    region_start = q2_regions[i]
    region_end = q2_regions[i+1]
    plt.axvline(x=region_start, color='gray', linestyle='--')
    plt.axvline(x=region_end, color='gray', linestyle='--')
    
    # Exclude regions 11 and 12 from labeling
    if i not in [4, 5]:
        plt.text((region_start + region_end) / 2, plt.ylim()[1] * 0.7, {i+1}, rotation=0, verticalalignment='center', horizontalalignment='center')


plt.legend()
#plt.grid(True)
plt.show()
```
